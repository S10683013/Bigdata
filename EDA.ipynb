```
library(tidyverse)
library(lubridate)
library(readxl)
```

```
abroad<- read_xlsx("歷年中華民國國民出國目的地人數統計.xlsx")
```

```
abroad
```

```
str(abroad)
```

```
#1.分析2002到2018年國民出境到大陸的人數比
abroad %>%
 count(年份,大陸) %>%
 ggplot(aes(x=年份, y=大陸,fill=年份))+
 geom_col()+
 geom_text(aes(label=大陸),vjust=-0.25)+
 labs(title="觀察2002到2018年國民出境到大陸的人數比",x="年份",y="人數")+theme(legend.position="none")
```

```
#2.分析韓國出國人數與年份關係
qplot(年份,韓國,data=abroad,geom=c("point","smooth"))
```

```
#3.分析總出國人數與年份之關係
abroad %>%
 count(年份,總計) %>%
 ggplot(aes(x=年份, y=總計,fill=年份))+
 geom_col()+
 geom_text(aes(label=總計),vjust=-0.25)+
 labs(title="分析總出國人數與年份之關係",x="年份",y="人數")+theme(legend.position="none")
```

```
#4.分析北美洲出國與美國出國總人數之關係
ggplot(abroad,aes(x=美洲合計,y=美國))+geom_smooth()+geom_point()
```

```
#5.分析總出國人數與亞洲出國總人數之關係
ggplot(abroad,aes(x=總計,y=亞洲合計))+geom_smooth()+geom_line()
```
