```
install.packages("SportsAnalytics")
library(SportsAnalytics)
NBA0910<-fetch_NBAPlayerStatistics("09-10")
#安裝並導入NBA_2009-2010年的資料
```

```
NBA0910order<-NBA0910[order(NBA0910$Assists,decreasing = T),]
NBA0910order[1:10,]
#用Assists作類別由大到小排出10名球員
```

```
subset(NBA0910,Position=="PG")
#從PG位置作欄位搜尋
```

```
NBA0910[grepl("Lebron",NBA0910$Name),]
#從名字搜尋含有Lebron的資料
```

```
install.packages("ggplot2")
library(ggplot2)
str(NBA0910)
```

```
qplot(TotalMinutesPlayed, TotalPoints, data=NBA0910)
#顯示x軸為"TotalMinutesPlayed"和y軸為"TotalPoints"的散佈圖
```

```
qplot(TotalMinutesPlayed, data=NBA0910, fill=Position)
#以TotalMinutesPlayed為主的直方圖
```

```
ggplot(NBA0910, aes(x=Position,y=TotalMinutesPlayed))+geom_boxplot()
#觀察Position在TotalMinutesPlayed的盒狀圖
```

```
ggplot(NBA0910, aes(x=TotalMinutesPlayed,y=TotalPoints))+geom_smooth()+geom_point()+facet_grid(Position~.)
#顯示x軸為"TotalMinutesPlayed"和y軸為"TotalPoints"的趨勢圖
#並以Potion做分類
```
