```
#導入2018-19的NBA資料
install.packages("ggplot2")
install.packages("SportsAnalytics")

library(SportsAnalytics)
library(ggplot2)

NBA1819<-fetch_NBAPlayerStatistics("18-19")
qplot(FieldGoalsAttempted, TotalPoints, data= NBA1819)
qplot(GamesPlayed, Steals, data= NBA1819)
```

```
qplot(GamesPlayed, data= NBA1516, fill=Position)
qplot(Steals, data= NBA1516, fill=Position)
```

```
ggplot(NBA1819,aes(x=FieldGoalsAttempted,y=TotalPoints))+geom_line()
ggplot(NBA1819,aes(x=GamesPlayed,y=Steals))+geom_line()
```

```
ggplot(NBA1819,aes(x=FieldGoalsAttempted,y=TotalPoints))+geom_smooth()
ggplot(NBA1819,aes(x=GamesPlayed,y=Steals))+geom_smooth()
```
