```
install.packages("readr")
install.packages("ggplot2")
library(readr)
library(ggplot2)
```

```
# 載入台南108年交通事故
tainandata <- read.csv("108.csv")
str(tainandata)
```

```
ggplot(tainandata, aes(x=當事者區分類別大類別名稱.車種, y=當事者事故發生時年齡))+geom_point()
```

```
ggplot(tainandata, aes(x=牌照種類名稱, y=當事者事故發生時年齡))+geom_boxplot()
```

```
ggplot(tainandata, aes(x=主要傷處名稱, y=當事者事故發生時年齡))+geom_boxplot()
```
